# Visualize Spectrums

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def visualize_spectrum(spectrum_df, num_peaks=None, max_draw=1, graph_label=None):
    """
    Visualize a dataframe containing spectrum data
    
    Parameters
      spectrum_df: pandas dataframe, each row is a channel
      num_peaks: true number of peaks in data (optional)
      max_draw: number of spectrums to draw (optional)
      graph_label: label this run (optional)
    """
    
    # Set figure size and resolution
    fig=plt.figure(figsize=(10, 5), dpi= 100)
    
    # Number of spectrums to draw on top of each other
    num_spectrums = min(max_draw, spectrum_df.shape[0])
    
    # Plot each channel
    for i in range(0, max_draw):
        plt.plot(spectrum_df.loc[i], linewidth=1)
    
    # Set plot title
    plt.title("Spectrum %s" % ("" if graph_label is None else "'"+graph_label+"'"))
    
    # Set plot subtitles
    text_skip = .06
    text_cursor = -.12
    
    if num_peaks is not None:
        plt.text(0, text_cursor, "Number of Peaks: %d" % num_peaks)
        text_cursor -= text_skip
        
    plt.text(0, text_cursor, "Number of Channels: %d" % max_draw)
    text_cursor -= text_skip
    
    # Plot formatting
    plt.tick_params(axis='both', which='both', bottom=False,
                    left=False, labelbottom=False, labelleft=False)
    
    # Show the plot
    plt.show()
```

```python
filename = "data.csv"

df = pd.read_csv(filename, skiprows=1, header=None)
num_peaks = int(pd.read_csv(filename, nrows=1, header=None)[0])

visualize_spectrum(df, num_peaks, max_draw=3)
```
